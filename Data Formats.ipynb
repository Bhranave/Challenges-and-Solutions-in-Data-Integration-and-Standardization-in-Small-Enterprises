```python
import pandas as pd
from datetime import datetime

# Define common date formats
date_formats = [
   "%Y-%m-%d",    # 2023-01-07
   # "%d-%m-%Y",    # 07-01-2023
   "%m/%d/%Y",    # 01/07/2023
    #"%d/%m/%Y",    # 07/01/2023
   # "%b %d, %Y",   # Jan 07, 2023
   # "%d %b %Y",    # 07 Jan 2023
]

# Read the CSV file
data = pd.read_csv('/content/drive/MyDrive/Thesis datasets/sales_data_sample - Copy.csv', encoding='latin1')

# Assuming the column with dates is named 'date_column'
date_column = data['Order Date']

# Function to identify the format of a date
def identify_date_format(date_str):
    for fmt in date_formats:
        try:
            datetime.strptime(date_str, fmt)
            return fmt  # Return the matching format
        except ValueError:
            continue
    return None  # Return None if no format matches

# Apply the function to the date column
data['detected_format'] = date_column.apply(identify_date_format)


# Loop through each detected date format
for fmt in date_formats:
    # Filter rows by the specific date format
    filtered_rows = data[data['detected_format'] == fmt]

    # Check if there are any rows with the current format
    if not filtered_rows.empty:
        # Get one sample (e.g., the first row)
        sample_row = filtered_rows[['Order No', 'Order Date']].iloc[0]

        # Print the format and the sample
        print(f"Sample for format {fmt}:")
        print(sample_row)
    else:
        # Print if no rows match the format
        print(f"No rows found with format {fmt}.")

```

```python
import pandas as pd

# Step 1: Load the CSV file
data = pd.read_csv('/content/drive/MyDrive/Thesis Notebooks/output_Problem 3_Lack of Standard Data Models (1).csv', encoding='latin1')

# Step 2: Convert 'Order Date' to datetime, let pandas infer the format
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

# Step 3: Reformat 'Order Date' to '%Y-%m-%d' only if it's not already in that format
# We will use a helper function to check the format
def reformat_date(date):
    if pd.isna(date):
        return date
    # Check if the date is already in '%Y-%m-%d' format
    if date.strftime('%Y/%d/%m') == date:
        return date
    return date.strftime('%Y/%d/%m')

# Apply the reformat_date function to the 'Order Date' column
data['Order Date'] = data['Order Date'].apply(reformat_date)

# Change 'Order Date' format into a new 'Order Date New' column
data['Order Date New'] = data['Order Date']

# Step 4: Save the updated DataFrame to a new CSV file
data.to_csv('updated_file.csv', index=False)

# Print the updated new data format
print(data[['Order No', 'Order Date New']])
```
