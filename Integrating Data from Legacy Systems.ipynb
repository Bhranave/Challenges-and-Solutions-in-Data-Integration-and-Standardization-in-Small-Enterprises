```python
import pandas as pd
from dateutil import parser
import json

# Step 1
# Load the CSV file
df = pd.read_csv('/content/drive/MyDrive/Thesis datasets/sales_data_sample - Copy.csv', encoding='latin1')

# Display the loaded data
print("Loaded CSV Data:\n", df)

# Step 2: Convert the Data to a JSON format
# Convert to a list of dictionaries for structured JSON output
json_data = df.to_dict(orient='records')

# Display the JSON data
print("\nConverted JSON Data:\n", json.dumps(json_data, indent=4))

# Step 3: Save the JSON data to a file format
json_file = 'modern_data.json'
with open(json_file, 'w') as f:
    json.dump(json_data, f, indent=4)

print(f"\nJSON file saved as '{json_file}'")
```
